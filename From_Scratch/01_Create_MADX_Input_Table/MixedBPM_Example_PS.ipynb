```python
# Thanks to G. Sterbini
def fromMatlabToDataFrame(listing, listOfVariableToAdd, verbose=False, matlabFullInfo=False):
    listOfVariableToAdd=list(set(listOfVariableToAdd))
    myDataFrame=pnd.DataFrame({})
    cycleStamp=[]
    cycleStampHuman=[]
    PLS_matlab=[]
    matlabObject=[]
    matlabFilePath=[]
    for j in listOfVariableToAdd:
        exec(j.replace('.','_')+'=[]')
    for i in listing:
        if verbose:
            print(i)
        data=japcMatlabImport(i);
        if matlabFullInfo:
            matlabObject.append(data)
        #to correct
        localCycleStamp=np.max(data.headerCycleStamps[1:]);
        deltaLocal_UTC=datetime.datetime.fromtimestamp(localCycleStamp/1e9)-datetime.datetime.utcfromtimestamp(localCycleStamp/1e9)
        utcCycleStamp=localCycleStamp+deltaLocal_UTC.total_seconds()*1e9
        cycleStamp.append(utcCycleStamp)
        aux=unixtime2datetimeVectorize(np.max(data.headerCycleStamps[1:])/1e9)
        cycleStampHuman.append(aux.tolist())
        PLS_matlab.append(data.cycleName)
        matlabFilePath.append(os.path.abspath(i))
        for j in listOfVariableToAdd:
            if hasattr(data,j.split('.')[0]):
                exec(j.replace('.','_') + '.append(data.' + j + ')')
            else:
                exec(j.replace('.','_') + '.append(np.nan)')
    myDataFrame['cycleStamp']=pnd.Series(cycleStamp,cycleStampHuman)
    myDataFrame['matlabPLS']=pnd.Series(PLS_matlab,cycleStampHuman)
    myDataFrame['matlabFilePath']=pnd.Series(matlabFilePath,cycleStampHuman)
    if matlabFullInfo:
        myDataFrame['matlabFullInfo']=pnd.Series(matlabObject,cycleStampHuman)
    for j in listOfVariableToAdd:
        exec('myDataFrame[\'' + j + '\']=pnd.Series(' +j.replace('.','_')+ ',cycleStampHuman)')    #myDataFrame=pnd.DataFrame({j:aux,
    return myDataFrame

def unixtime2datetimeVectorize(x):
        """Transform unixtime in python datetime"""
        aux=np.vectorize(datetime.datetime.fromtimestamp)
        if np.size(x)!=0:
            return aux(x)
        else:
            return []
        
def japcMatlabImport(myfile):
    """Import the in a python structure a matlab file""" 
    myDataStruct = loadmat(myfile,squeeze_me=True, struct_as_record=False)
    return myDataStruct['myDataStruct']

def fromTimberToDataFrame(listOfVariableToAdd,t1,t2,verbose=False, fundamental=''):
    listOfVariableToAdd=list(set(listOfVariableToAdd))
    if fundamental=='':
        DATA=log.get(listOfVariableToAdd,t1,t2 )
    else:
        DATA=log.get(listOfVariableToAdd,t1,t2,fundamental)
    myDataFrame=pnd.DataFrame({})
    if DATA!={}:
        for i in listOfVariableToAdd:
            myDataFrame[i]=pnd.Series(DATA[i][1].tolist(),unixtime2datetimeVectorize(DATA[i][0]))
        myDataFrame['cycleStamp']=np.double(myDataFrame.index.astype(np.int64))
    return myDataFrame

```

```python
def movingAverage(mydata,window=5):
    ''' The movingAverage function returns a time series after subtracting its moving average. 
    The format is movingAverage(mydata,window=5,periods=1), where mydata is the input NumPy array, 
    window is the stensile size and periods is the minimum data without moving average. 
    The last samples equal to the number of window minus 1, are substituted by the value minus the normal average. The output is again a NumPy array.'''
    
    mydata=mydata[::-1]
    mydata=pnd.DataFrame(mydata)
    
    temp=mydata.rolling(window,center=False).mean()
    temp=temp[::-1]
    temp[-window:]=np.tile(np.mean(temp[-window:]),(window,1))
    output=mydata[::-1]-temp
    output=output[1::]# We take out the first turn because the closing bump causes an overshoot due to the normal mean operation above. TODO: One might actually discard the first window-1 number of data.
#     output[0,:]=np.nanmean(output) # Due to the issue with the first turn, put the average of the signal as first value..Probably needs to be optimized.
    return output.values


def read_PS_data(mydir,bunch,myfields='PR_BPM.AcquisitionTrajectoryBBB.value.position',mean=1,limit=10,win=5,mixed=0,svd=0,mode=6,N1=1,N2=300):
    ''' A function to automatically read MATLAB turn by turn data from the PS. First, the data are extracted with the functions
        from Guido's toolbox into a Panda's dataframe for the bunch which the user requested. The function returns the data in a dataframe. 
        
        Input:
        
        mydir: Directory with the .mat files. Do not add '/' at the end.
        bunch: Number of bunch in the PS
        myfields: The field name of the TbT data.
        mean: if True, moving average is employed to remove the DC part of the signal. Set to True for frequency analysis.
        limit: The total number of .mat files to be included.
        win: The moving average window
        mixed: Flag to create a field with an interleaved BPM signal for mixed BPM analysis.
        svd: Flag to activate Singular Value Decomposition (SVD) analysis. Removes noise based on the mode number. Creates an extra field in the final result.
        mode: How many modes to keep in the SVD filtering.
        N1: The first turn in the range of turns to be extracted.
        N2: The last turn in the range of turns to be extracted.
        
        Output:
        
        Datafram with timestamps of the chosen .mat files and the beam position, together with the extra fields generated with respect to the input from the user.
        
        '''
    
    mydir=mydir+'/*.mat'
    turns=N2-N1+1
    myFiles=sorted(glob.glob(mydir))
    myDataFormat=japcMatlabImport(myFiles[0])
    myaux=list(np.unique(list(myDataFormat.PR_BPM.AcquisitionTrajectoryBBB.value.bucketId)))
    
    myind=myaux.index(bunch)
    mylen=len(myaux)
    
    myexp=myDataFormat.PR_BPM.AcquisitionTrajectoryBBB.value.position_unitExponent
    
    auxh=np.hstack([np.arange(17,43),np.arange(0,17)])
    auxv=np.hstack([np.arange(17+43,43+43),np.arange(0+43,43+17)])
    
    myDataFrame=fromMatlabToDataFrame(myFiles[0:limit],[myfields])
    
    myDataFrame['Position H']=myDataFrame['PR_BPM.AcquisitionTrajectoryBBB.value.position'].\
    apply(lambda x: np.transpose(np.double(x[auxh,myind::mylen])*(10**myexp))[N1-1:N2,:])
    myDataFrame['Position V']=myDataFrame['PR_BPM.AcquisitionTrajectoryBBB.value.position'].\
    apply(lambda x: np.transpose(np.double(x[auxv,myind::mylen])*(10**myexp))[N1-1:N2,:])

    if mean:
        myDataFrame['Position H NoDC']=myDataFrame['Position H'].apply(lambda x: movingAverage(x,window=win))
        myDataFrame['Position V NoDC']=myDataFrame['Position V'].apply(lambda x: movingAverage(x,window=win))
    if svd:
        myDataFrame['SVD H']=myDataFrame['Position H NoDC'].apply(lambda x: my_svd(x,mode))
        myDataFrame['SVD V']=myDataFrame['Position V NoDC'].apply(lambda x: my_svd(x,mode))
    if mixed:
        myDataFrame['Mixed Position H']=myDataFrame['Position H NoDC'].apply(lambda x: np.ndarray.flatten(x))
        myDataFrame['Mixed Position V']=myDataFrame['Position V NoDC'].apply(lambda x: np.ndarray.flatten(x))
    if mixed and svd: 
        myDataFrame['Mixed SVD H']=myDataFrame['SVD H'].apply(lambda x: np.ndarray.flatten(x))
        myDataFrame['Mixed SVD V']=myDataFrame['SVD V'].apply(lambda x: np.ndarray.flatten(x))
        
    print('Processing of MATLAB data finished. Total of: Data consist of {0} turns X {1} BPMs.'.format(turns,len(auxh)))    
    return myDataFrame

def my_svd(data,mode=0):
    u,s,v=np.linalg.svd(data,full_matrices=False)
    if not mode:
        fig=plt.figure(figsize=(10,5))
        plt.stem(s/max(s));plt.grid('on');plt.xlabel('Mode');plt.ylabel('Normalized Eigenvalues')
        plt.show()
        mode=input("Where would you like to set the cut-off mode? For example, type 2 to keep the first two modes.")
        plt.close(fig)
    
    s[mode:]=0
    
    return np.dot(u,np.dot(np.diag(s),v))
```

## Functions for frequency analysis with PyNAFF

```python
def freq_analysis(data,N_init,N_step,N_final,mixed=False,M=0):
    
    ''' Function for frequency analysis, with a single BPM or by mixing all the BPMs.
    
     data: Array of the data in (N,M) format, where N are the turns, M are the BPMs.
    
     N_init: Initial turn of the analysis.
    
     N_step: Step of the sliding window.
    
     N_final: Final turn of the analysis.
    
     mixed: Flag for triggering mixed BPM* analysis (True) or single BPM analysis (False).
    
     M: In case of single BPM, the analysis is performed on the data of the M BPM.
    
     Output: The vector with the tune estimations, and the vector with the number of turns used in the analysis.
    
     *Info on the mixed BPM method: https://journals.aps.org/prab/abstract/10.1103/PhysRevAccelBeams.22.071002 '''
    
    scan_turns=np.linspace(N_init,N_final-N_step,N_final-N_step+(1-N_init),dtype=int)
    
    q=np.empty(len(scan_turns))
    
    if mixed:
        aux=np.ndarray.flatten(data)
        fac=np.shape(data)[1]
    else:
        aux=data[:,M]
        fac=1
    
    for m,n in zip(range(len(scan_turns)),scan_turns):
        
        y=aux[fac*n:fac*(n+N_step+10)]
        try:
            q[m]=abs(pnf.naff(y,turns=fac*N_step,nterms=1,
                                    skipTurns=0,getFullSpectrum=1,window=1)[0][1])*fac
            
        except:

            q[m]=np.nan
    
    if mixed:
        q=q%(1/2)
        
    return q,scan_turns
```

## Instantaneous frequency analysis with single and Mixed BPM analysis, by employing a sliding window

```python
N_init=60
N_step=24
N_final=1500

my_turns=np.linspace(N_init,N_final-N_step,N_final-N_step+(1-N_init),dtype=int)
q_single_all=np.empty((len(myData.index),len(my_turns)))

q_mixed_all=np.empty((len(myData.index),len(my_turns)))

# I study all the 5 injections here to compute the average, remove the for loop if you wish to study one case only.
for k in range(len(myData.index)):
                 
    x1=myData['Position H NoDC'].iloc[k].copy() 
    
    q_single_all[k,:],_=freq_analysis(data=x1,N_init=N_init,N_step=N_step,N_final=N_final,M=0,mixed=False)

    q_mixed_all[k,:],_=freq_analysis(data=x1,N_init=N_init,N_step=N_step,N_final=N_final,mixed=True)
    

    
```

## Show tune periodic variation while the bumpers are fired up from minimum to maximum field, for both mixed and single BPM analyses.
